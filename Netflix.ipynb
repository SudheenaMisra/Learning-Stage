```python
# Create the years and durations lists
years = [2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]
durations = [103,101,99,100,100,95,95,96,93,90]

# Create a dictionary with the two lists
movie_dict = dict(zip(years,durations))

# Print the dictionary
movie_dict
```

```python
# Import pandas under its usual alias
import pandas as pd

# Create a DataFrame from the dictionary
durations_df = pd.DataFrame.from_dict(movie_dict, orient='index')

# Print the DataFrame
durations_df
```

```python
# Import matplotlib.pyplot under its usual alias and create a figure
import matplotlib.pyplot as plt
fig = plt.figure()

# Draw a line plot of release_years and durations
plt.plot(years, durations)

# Create a title
plt.title("Netflix Movie Durations 2011-2020")

# Show the plot
plt.show()
```

```python
netflix_df = pd.read_csv("https://raw.githubusercontent.com/SudheenaMisra/Datasets/main/netflix_data.csv")
netflix_df.head()
```

```python
# Subset the DataFrame for type "Movie"
netflix_df_movies_only = netflix_df[netflix_df["type"]=='Movie']

# Select only the columns of interest
netflix_df_subset_col = netflix_df[['title','country','release_year','duration','genre']]

# Print the first five rows of the new DataFrame
netflix_df_subset_col.head()
```

```python
# Create a figure and increase the figure size
plt.figure(figsize=(12,15))

# Create a scatter plot of duration versus year
plt.scatter(netflix_df_subset_col[['release_year']],netflix_df_subset_col[['duration']])

# Create a title
plt.title("Movie duration by release year")

# Show the plot
plt.show()
```

```python
# Filter for durations shorter than 60 minutes
short_movies = netflix_df_subset_col[netflix_df_subset_col['duration']<60]

# Print the first 20 rows of short_movies
print(short_movies.head(n=20))
```

```python
# Define an empty list
colors = []

# Iterate over rows of netflix_movies_col_subset
for lab,row in netflix_df_subset_col.iterrows():
  if row['genre'] =="Children":
    colors.append("red")
  elif row['genre'] == "Documentaries":
    colors.append('blue')  
  elif row['genre'] == "Stand-Up":
    colors.append('green')
  else:
    colors.append('black')
    
# Inspect the first 10 values in your list  
colors[0:10]
```

```python
# Set the figure style and initalize a new figure
plt.style.use("fivethirtyeight")
fig = plt.figure(figsize=(12,8))
# Create a scatter plot of duration versus release_year with color on each genre
plt.scatter(x=netflix_df_subset_col[['release_year']],y=netflix_df_subset_col[["duration"]], c=colors)

#adding title
plt.title("Movie duration by release Year")
#Adding axis labels
plt.xlabel("Release Year")
plt.ylabel("Duration")

#showing figure
plt.show()
```
